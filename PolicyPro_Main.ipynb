```python
# ======================================
# STEP 1 – Install Requirements (Run Once)
# ======================================

!pip install spacy scikit-learn pymupdf streamlit pandas
!python -m spacy download en_core_web_sm


# ======================================
# CELL 1 – Import Libraries
# ======================================

import re
import fitz
import spacy
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

print("Libraries Imported")


# ======================================
# CELL 2 – Load spaCy Model
# ======================================

nlp = spacy.load("en_core_web_sm")

print("spaCy Model Loaded")


# ======================================
# CELL 3 – Extract Text From PDF
# ======================================

def extract_text_from_pdf(pdf_file):

    text = ""

    doc = fitz.open(stream=pdf_file.read(), filetype="pdf")

    for page in doc:
        text += page.get_text()

    return text


# ======================================
# CELL 4 – Extract Text From TXT
# ======================================

def extract_text_from_txt(txt_file):

    return txt_file.read().decode("utf-8")


# ======================================
# CELL 5 – Clean Text
# ======================================

def clean_text(text):

    text = text.lower()

    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    text = re.sub(r'\s+', ' ', text)

    return text.strip()


# ======================================
# CELL 6 – Sentence Tokenization
# ======================================

def sentence_tokenize(text):

    doc = nlp(text)

    sentences = [sent.text.strip() for sent in doc.sents]

    return sentences


# ======================================
# CELL 7 – TF-IDF Summarization
# ======================================

def tfidf_summarize(sentences, top_n=5, word_limit=200):

    vectorizer = TfidfVectorizer(stop_words="english")

    tfidf_matrix = vectorizer.fit_transform(sentences)

    sentence_scores = np.sum(tfidf_matrix.toarray(), axis=1)

    ranked = np.argsort(sentence_scores)[-top_n:]

    selected = [sentences[i] for i in sorted(ranked)]

    summary = " ".join(selected)

    words = summary.split()

    summary = " ".join(words[:word_limit])

    return summary


# ======================================
# CELL 8 – Entity Extraction
# ======================================

def extract_entities(text):

    doc = nlp(text)

    entities = [(ent.text, ent.label_) for ent in doc.ents]

    return entities


# ======================================
# CELL 9 – Keyword Extraction
# ======================================

def extract_keywords(text, top_n=10):

    vectorizer = TfidfVectorizer(stop_words="english", max_features=top_n)

    vectorizer.fit([text])

    keywords = vectorizer.get_feature_names_out()

    return keywords


print("All Backend Functions Ready")
```
